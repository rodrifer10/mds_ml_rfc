## Clave del SVM: TIENE QUE SER LINEALMENTE SEPARABLE EL PROBLEMA, sino no se puede!
Pero que pasa, que con muchas variables es casi imposible saber si es linealmente separable o no.. jajja

Entonces bueno, ver que se hace...

Una línea de 3 dimensiones en un plano, en 4 dimensiones son figuras geométicas, dsps ya pasamos a Hiperplanos...

**Ojo, el SVM NO genera curvas! Otros sistemas podrían servir pero NO SVM!**

### La idea es encontrar no solo la mejor recta, sino también los mejores márgenes (soportes!), más amplios y mejores posicionados de error!

Cada combinación posible 2a2 de variables me da un soporte, la combinación de todos los soportes, osea de todas las combinaciones de cada variable con el target es la solución! Cada combinación con un peso relativo siempre...

Este peso, osea COEFICIENTE es un parámetro que el mismo modelo ESTIMA SOLO, por eso es tan pesado!

### Loss-Function del SVM:

Tiene una parte similar a la logística! --> Ver fórmula en Slides, es importante!

Pero además de esa loss, tiene un **Factor de regularización** que va optimizando la función!

||w|| --> módulo de un vector (buscar)

**Clave conceptual: Lo que busca la función es EL VECTOR (FUNCIÓN) MÁS SIMPLE QUE EXPLIQUE el modelo, optimizando (minimizando) el error!**

||w||^2 --> El menor módulo significa el vector más simple! Error menor y facilita más los cálculos!

* C dentro de la fórmula es un parámetro que pondera al error!
    - C = infinito --> Cualquier error distinto a 0 tengo un error gigante y no encuentro la solución --> Soy super exigente al buscar! Pq castigo muchísimo los errores
    - C alto --> MÁRGENES MÁS ESTRECHOS! Más precisión digamos... --> Necesito más potencia computacional pq necesito más iteraciones... riesgo de nunca conseguirlo...
        - Como el cualquier técnica, mientras más precisión --> peligro de OVERFITTING
    - C bajo --> MÁRGENES MAYORES! Menor precisión!
Entonces: Mientras más alto el C, MÁS PENALIZO EL ERROR, OSEA MENOR ERROR ADMITO!

# KERNEL --> Concepto CLAVE en matemática y álgebra... Y EN MACHINE LEARNING

Es el valor de la variable que hace a la función igual a 0 --> Algo así sería el Kernel

Como puedo convertir un problema que NO es linealmente separable en principio, en otro que SI lo es

Es como inventarme otra dimención, una coordenada Z a una función x,y. Es una función que aplicada a un x,y me devuelve un x,y,z.

**Tiene que hacerce con cierta lógica!** --> No se puede aplicar funciones de Kernel a cualquier cosa!

Diseñar una función de Kernel es bien complicado, NO vale meter cualquier cosa!!!

* Hay un listado de funciones de Kernel para X casos... Tienen bastantes limitaciones que hay q investigar...
* El modelo NO encuentra funciones de kernel solas... Hay que buscar una a aplicar entre las que están leyendo la documentación!

**SIEMPRE TENGO QUE METER UNA FUNCIÓN DE KERNEL! PQ NO TENGO GARANTÍAS DE QUE SEA LINEALMENTE SEPARABLE. PARA ASEGURARME TENGO QUE METER FUNCIÓN DE KERNEL!**
