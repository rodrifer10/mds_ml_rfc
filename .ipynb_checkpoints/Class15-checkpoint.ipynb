## K-Means

La idea es que la varianza DENTRO del cluster sea muy BAJA, pero que ENTRE CLUSTERS sea muy ALTA.

1. **ELIJO LA CANTIDAD DE CLUSTERS (Parámetro de entrada base) --> Ver metodologías!**
2. Elige centroides aleatoriamente (cualq punto) en la primer iteracion
3. Mide la DISTANCIA EUCLIDEA (siempre rectas de distancia, no curvas) de cada punto al centroide
4. Asigna esos puntos al cluster del centroide al que tenga menor distancia
5. Se vuelven a elegir centroides, buscando esta vez el punto central de cada cluster!
6. Se vuelven a calcular las distancias y las instancias pueden cambiar de cluster
7. Se itera así... hasta cuando?:
    - Criterio de n iteraciones, pongo un nro de iteraciones --> SKLearn hace este
    - Cuando de una iteración a otra, ningún punto cambia de cluster... Pero esto puede ser muy pesado computacionalmente!
    - Cuando haya menos de X distancia al centroide

## DBSCAN

Clusteriza en función de DENSIDADES. Es decir, por las modas.

Puntos muy juntos van al mismo cluster, pero puntos muy separados NO VAN A NINGÚN CLUSTER!

Osea, deja afuera a los outliers y no los clasifica (les pone -1), muy bueno! -->Los clasifica como ruido en realidad...

**Esto está muy bueno para meterle a la práctica, porque detecta OUTLIERS POR FILAS (nosotros detectamos x filas). Entonces me quedan marcados con -1 los puntos muy separados, osea los anómalos!**

AL FIJARSE EN LAS DENSIDADES Y QUE TAN CERCANOS ESTÁN LOS PUNTOS, ME CLASIFICA BIEN COSAS COMO EL MOONS!!

* Es mejor cuando NO hay centroides fijos!
* Cuando la distancia NO sea lineal nomás, sino curva

# Parámetros:

* **minPts**: Número mínimo de instancias en cada Cluster! --> ojo, que si es muy alto no clasificará nada en un cluster (o todos si están muy cerca)
* **Epsilon**: Distancia máxima que puede haber entre 2 puntos de un cluster! --> Dejo muchos outliers y los puntos juntitos los agrupa

* SOLO TIENE FIT_PREDICT! --> No se puede hacer .fit x un lado y .predict en otro!
