### Undersampling y Oversampling:

Ambos tratan de Balancear (no siemore 50 y 50, aveces 70-30 o depende) el DataSet!!!

1. Undersampling: Trato de seleccionar los datos en una muestra (y que sea muy representativa del total de los datos!) para achicar el dataset y que haya un % más similar de datos entre los 0 y 1

# Otra cosa CLAVE: PRECISION Y RECALL

Ambos se calculan SOBRE UNA CLASE A LA VEZ!

1. **Precision**: Cuantos items de los seleccionados son RELEVANTES. Especificidad, osea es el ratio de negativos por así decirlo...
    - Tiene MUCHO en cuenta cuánto acierto y cuanto la cago!
    - Si la hago con los negativos se llamaría especificidad (specificy)...
    - Si la precisión es MALA (BAJA), quiere decir q los falsos POSITIVOS son ALTOS!!!

2. **Recall (Sensibility tmb llamado)**: A CUANTOS CAPTURO (cuantos de los que son fraude detecté como fraudes?). **UN MODELO BASE TIENE 0% DE RECALL!!!** (o del 100% si lo calculo sobre la clase mayoritaria... pero eso no sería lo que busco)
    - Si el recall es muy bajo, tomo mucho fraude, por lo q los falsos negativos van a ser muy altos...
    - Si me sale 0, no pillo a ninguno! Entonces, si yo digo q todos son 0, voy a tener MUCHOS FALSOS NEGATIVOS!

Los puedo calcular sobre CUALQUIER clase!

**El F1 score viene a ser una combinación de ambos..**


# Curva ROC:

Una Confussion Matrix se arma con 0 y 1, pero muchas veces yo tengo probabilidades. Para forzarlas tengo que marcar un UMBRAL para targetear como 0 y 1 a los resultados de la proba...

En el fondo, cuando trabajo con probas, no tengo una Matriz de confusión, sino **INFINITAS MATRICES DE CONFUSIÓN EN BASE AL UMBRAL!!!**

Esas infinitas combinaciones de Accuracy, Precision y Recall en base a CADA UMBRAL de la confussion Matrix

TPR (TRUE POSITIVE RATE) - Es lo mismo que el Recall para los Positives!

FPR (FALSE POSITIVE RATE) - Es el recall para la otra variable, para los Negatives en este caso!

CON ESTOS DOS PARÁMETROS HAGO LOS EJES!!! Osea que genero infinitas matrices de confusión con ambos Ratios!

**El mejor modelo es el que cubre LA MAYOR SUPERFICIE POR DEBAJO DE LA CURVA! (con integrales la podemos sacar, la integral de 0 a 1 pq son probabilidades)**



http://www.chioka.in/differences-between-roc-auc-and-pr-auc/


### OJO, ESTO NO TIENE EN CUENTA EL COSTO NI EL ACCURACY! Hay que mirar TODO lo que tengamos y sea relevante al analizar un modelo!

Nunca la ROC puede ser menor a 0.5! Mientras más cerca a la 0.5 es PEOR!

# F SCORES:

F1, F0.5, F2... el número ES LA BETA! (un parámetro de la métrica!)

Está en base a la precision y al recall!

* beta 1 = armonic mean entre precision y recall
* beta > 1 favors recall
* beta < 1 favors precision

Ver fórmula: https://en.wikipedia.org/wiki/F-score

**En función de la importancia que le quiero dar al Recall o a la Precision, voy a ir tocando la Beta!**

Por CONVENCIÓN se usan 0.5 y 2, puede ser modificable pero x lo gral se usan ese tipo de Betas...

**El que elijamos va a depender de lo que busque con mi modelo y el tipo de negocio que vaya buscando...**

No son contrarias Precision y Recall, pero por lo general ELEVAR UNO DISMINUYE EL OTRO...
