## Descripción de Variables

#### El artículo descriptivo del DataFrame describe a cada una de las variables que lo componen de la siguiente manera:

* fraud_bool (*binary*): If the application is fraudulent or not. **Variable Target**. 0 si la transacción NO es fraudulenta, 1 si lo es.
* income (numeric): Annual income of the applicant (in decile form). Ranges between [0.1, 0.9].
* name_email_similarity (numeric): Metric of similarity between email and applicant’s name. Higher values represent higher similarity. Ranges between [0, 1].
* prev_address_months_count (numeric): Number of months in previous registered address of the applicant, i.e. the applicant’s previous residence, if applicable. Ranges between [−1, 380] months **(-1 is a missing value)**.
* current_address_months_count (numeric): Months in currently registered address of the applicant. Ranges between [−1, 429] months **(-1 is a missing value)**
* customer_age (numeric): Applicant’s age in years, rounded to the decade. Ranges between [10, 90] years.
* days_since_request (numeric): Number of days passed since application was done. Ranges between [0, 79] days.
* intended_balcon_amount (numeric): Initial transferred amount for application. Ranges between [−16, 114] **(negatives are missing values).**
* payment_type (categorical): Credit payment plan type. 5 possible (annonymized) values.
* zip_count_4w (numeric): Number of applications within same zip code in last 4 weeks. Ranges between [1, 6830].
* velocity_6h (numeric): Velocity of total applications made in last 6 hours i.e., average number of applications per hour in the last 6 hours. Ranges between [−175, 16818].
* velocity_24h (numeric): Velocity of total applications made in last 24 hours i.e., average number of applications per hour in the last 24 hours. Ranges between [1297, 9586]
* velocity_4w (numeric): Velocity of total applications made in last 4 weeks, i.e., average number of applications per hour in the last 4 weeks. Ranges between [2825, 7020].
* bank_branch_count_8w (numeric): Number of total applications in the selected bank branch in last 8 weeks. Ranges between [0, 2404].
* date_of_birth_distinct_emails_4w (numeric): Number of emails for applicants with same date of birth in last 4 weeks. Ranges between [0, 39].
* employment_status (categorical): Employment status of the applicant. 7 possible (annonymized) values.
* credit_risk_score (numeric): Internal score of application risk. Ranges between [−191, 389].
* email_is_free (*binary*): Domain of application email (either free or paid).
* housing_status (categorical): Current residential status for applicant. 7 possible (annonymized) values.
* phone_home_valid (*binary*): Validity of provided home phone.
* phone_mobile_valid (*binary*): Validity of provided mobile phone.
* bank_months_count (numeric): How old is previous account (if held) in months. Ranges between [−1, 32] months **(-1 is a missing value).**
* has_other_cards (*binary*): If applicant has other cards from the same banking company.
* proposed_credit_limit (numeric): Applicant’s proposed credit limit. Ranges between [200, 2000].
* foreign_request (*binary*): If origin country of request is different from bank’s country.
* source (categorical): Online source of application. Either browser (INTERNET) or app (TELEAPP).
* session_length_in_minutes (numeric): Length of user session in banking website in minutes. Ranges between [−1, 107] minutes **(-1 is a missing value).**
* device_os (categorical): Operative system of device that made request. Possible values are: Windows, macOS, Linux, X11, or other.
* keep_alive_session (*binary*): User option on session logout.
* device_distinct_emails (numeric): Number of distinct emails in banking website from the used device in last 8 weeks. Ranges between [−1, 2] emails **(-1 is a missing value).**
* device_fraud_count (numeric): Number of fraudulent applications with used device. Ranges between [0, 1].
* month (numeric): Month where the application was made. Ranges between [0, 7].

## Tratamiento de Missing Values:

#### De la descipción de variables por parte de los autores surgen de las siguientes descripciones la existencia de Missing Values codificados en algunos atributos del DataSet. Por ende las variables Missing a revisar serán las sigueintes:
* prev_address_months_count (numeric): Number of months in previous registered address of the applicant, i.e. the applicant’s previous residence, if applicable. Ranges between [−1, 380] months **(-1 is a missing value)**.
* current_address_months_count (numeric): Months in currently registered address of the applicant. Ranges between [−1, 429] months **(-1 is a missing value)**
* intended_balcon_amount (numeric): Initial transferred amount for application. Ranges between [−16, 114] **(negatives are missing values).**
* bank_months_count (numeric): How old is previous account (if held) in months. Ranges between [−1, 32] months **(-1 is a missing value).**
* session_length_in_minutes (numeric): Length of user session in banking website in minutes. Ranges between [−1, 107] minutes **(-1 is a missing value).**
* device_distinct_emails (numeric): Number of distinct emails in banking website from the used device in last 8 weeks. Ranges between [−1, 2] emails **(-1 is a missing value).**

En muchas variables, más que nada en las numéricas (también en alguna categórica), los valores en negativo son MISSING, por lo que hay que determinar como los vamos a tratar

* prev_address_months_count
* current_address_months_count
* intended_balcon_amount
* bank_months_count
* session_length_in_minutes
* device_distinct_emails


```python
df_fraud_cat, df_fraud_num
```

Tenemos 4 variables numéricas con missings y 2 categóricas

```python
cols_with_mis = ['prev_address_months_count'
            ,'current_address_months_count'
            ,'intended_balcon_amount'
            ,'bank_months_count'
            ,'session_length_in_minutes'
            ,'device_distinct_emails_8w']
```

```python
#for i in cols_with_mis:
#    print(i, " : ", var_description.set_index('Variable').loc[i,'Description'])

fr.dame_info(cols_with_mis, df_fraud, var_description, only_desc=True)
# Veo la descripción de cada variable para volver a revisar de que se compone cada una y analizar los missings en contexto
```

```python
df_fraud[cols_with_mis][df_fraud[cols_with_mis] < 0].count()/df_fraud.shape[0]
```

Bueno, lo que en principio parecían ser pocos missings, ahora parece ser que hablamos de columnas con muchos datos de este tipo.

Las más preocupantes son claramente prev_adress_months_count y intended_balcon_amount



```python
df_fraud[df_fraud['prev_address_months_count'] < 0].fraud_bool.value_counts(normalize=True)
```

```python
df_fraud[df_fraud['prev_address_months_count'] > 0].fraud_bool.value_counts(normalize=True)
```

Aunque la tendencia sea relativamente similar, se puede notar una diferencia entre lo que pasa cuando los valores son missing o no, es decir que, a pesar de ser muchos los missings, pueden darme algún tipo de información

```python
df_fraud[df_fraud['intended_balcon_amount'] < 0].fraud_bool.value_counts(normalize=True)
```

```python
df_fraud[df_fraud['intended_balcon_amount'] > 0].fraud_bool.value_counts(normalize=True)
```

Lo mismo pasa con esta columna! Vemos que, aunque parezca poca la diferencia, en realidad hablamos de que cuando el valor es missing, las chances de que haya fraude mucho más altas comparándolo vs cuando la variable no tiene un valor missing. Esta información puede ser útil, por lo que debería mantenerla.

```python
df_fraud[df_fraud['intended_balcon_amount'] < 0].fraud_bool.value_counts(normalize=True) / df_fraud[df_fraud['intended_balcon_amount'] > 0].fraud_bool.value_counts(normalize=True) -1
```

### En conclusión, sobre las variables que tienen muchos Missing Values:

Creo que no amerita una transformación (ya que el missing ya está correctamente codificado, al ser negativo es bien diferente del resto de los valores y no es un Null) y tampoco amerita sacar la columna del análisís, ya que nos puede brindar información. Solo hay dos variables que rondan el 70% y en el resto los porcentajes son bajos, ninguna ronda el 90% de valores missings, porcentaje que ya nos haría evaluar el eliminar dicha variable del modelo.

Cabe destacar que esto además me parece lógico debido al tipo de proyecto que estamos tratando. En otros modelos la falta de datos puede ser simplemente falta de información, pero en este caso de aplicaciones a cuentas bancarias y detección de fraude, el que no exista información, ya sea porque el cliente no la ha cargado, porque el proveedor de datos no posee información de ese cliente o por cualquier razón, nos puede indicar algo sobre el potencial del fraude. Es decir, en este caso particular de análisis el hecho de no poseer información sobre alguien puede ser información útil también.

Por ende, se mantendrán las variables con missings y su codificación actual ya que es útil así. De igual manera, a la hora de ejecutar los modelos será un aspecto a tener en cuenta, para ver si afectan de manera negativa o no.
