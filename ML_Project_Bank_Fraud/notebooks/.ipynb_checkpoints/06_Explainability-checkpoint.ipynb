# 2. Gráficos de dependecia parcial

```python
%%time
shap.partial_dependence_plot(
    "credit_risk_score",
    lgbm.predict,
    data=X_test,
    ice=False,
    model_expected_value=True,
    feature_expected_value=True,
)
```

Este tipo de gráficos son especialmente interesantes debido a que nos permiten ver que características de determinadas variables pueden generar una mayor
o menor probabilidad de fraude en nuestro modelo. Y esta variable de score de riesgo crediticio explica de gran manera dicha relación.

Se puede observar como a medida que el score de riesgo crediticio aumenta, especialmente por encima de 200, se dispara la influencia de la variable en el modelo de manera abrupta. Esto nos indica que valores altos tendrán un mayor peso en la probabilidad que asignará el modelo a la instancia, en este caso seguramente aumentando las probabilidades. Esto es muy interesante.

Además se le suma información valiosa como la distribución en gris y la mediana como línea desde el eje X, pero en especial la línea horizontal que nos indica el valor normal esperado de la variable en el modelo. Podemos decir que valores por encima de este, como lo son a partir de los 200, tendrán un impacto importante superior al promedio de las instancias.

```python
shap.dependence_plot('credit_risk_score', shap_values[1], X_test)
```

Además del gráfico de dependencias parciales, la librería también ofrece gráficos de dependencia de este tipo, en donde relaciona el valor SHAP de una
variable con sus valores en general y además discrimina por otra variable con la que exista cierto grado de correlación por lo que se entiende.

En este caso es muy interesante ver como, a medida que crece el score crediticio y por ende su valor shap (algo que vimos en el gráfico anterior), también suelen surgir valores más altos del límite de crédito propuesto por el usuario. Esto se torna aún más interesante cuando recordamos que en el EDA ya notamos una correlación mediana entre estas variables y este tipo de relación positiva, cosa que luego de aplicar el modelo se puede notar claramente también.

Parece que el valor 0 toma una gran preponderancia para el modelo cuando aparece y veremos que aumenta claramente las probabilidades de que exista fraude en este sentido.

## 5. Acercamiento al Clustering mediante shap values

```python
shap_df = pd.DataFrame(shap_values[1], columns=X_test.columns.values)
```

```python
shap_4cluster = pd.concat([shap_df, y_test], axis=1)
shap_4cluster
```

```python
shap_4cluster.to_csv('../data/processed/shap_values_df_test.csv', index=False)
```

```python
#shap_4cluster = pd.read_csv('../data/processed/shap_values_df_test')
```

```python
shap1 = shap_4cluster[shap_4cluster['fraud_bool']==1]
```

```python
%%time
fr.k_means_search(shap1,10)
```

```python
km = KMeans(n_clusters=3)
km.fit(shap1)
```

```python
km_coefs = km.transform(shap1)
```

```python
cluster = []
for i in km_coefs:
    cluster.append(i.argmax())
```

```python
km_coefs
```

```python
len(cluster)
```

```python
km_coefs.shape
```

```python
shap1['cluster'] = np.array(cluster) + 1 # para que el nro de clusters vaya del 1 al 3 en vez del 0 al 2
```

```python
frauds = df_fraud_test[df_fraud_test.fraud_bool==1]
frauds['cluster'] = np.array(cluster) + 1
```

```python
bools, cats, nums = fr.tipos_vars(df=frauds)
```

```python
frauds.income = round(frauds.income,4)
```

```python
frauds.cluster.value_counts()
```

La idea con este tipo de clusterización sería el realizar un análisis descriptivo que nos permita identificar patrones por los que se agrupan los usuarios
que cometen fraude y tomar decisiones ya sea de prevensión, detección o reacción frente a dichas conductas.
