# <font size=25><b>Conclusiones finales</b></font>

Una vez finalizados todos los módulos de este proyecto, en el que hemos ocupado todas las fases del proceso de generación de un modelo de machine learning para la detección de fraude en aplicaciones para apertura de cuentas bancarias, procederé a brindar un pequeño resumen que recapitule los hallazgos y conclusiones más importantes que han sido recavadas durante todo el análisis.

En primer lugar, volveremos a comentar lo dicho en un principio, en cuanto a que el objetivo de este trabajo se ha centrado principalmente en aprender todo lo posible sobre el proceso de aplicación de modelos predictivos de machine learning sobre un dataset lo más parecido posible a la realidad. Esta última afirmación puede ser debatible, debido a que, como hemos visto, el proceso de codificación y anonimización realizado al dataset para poder publicarlo limita ciertos razgos de nuestro análisis, además de que carecemos de información relevante desde el punto de vista de un negocio real, lo que hubiese nutrido mucho más al proyecto, pero a pesar de esto el objetivo puede considerarse bastante cumplido. Esto debido a que a lo largo del proyecto hemos ido pasando por cada una de las etapas de manera exitosa, logrando aplicar las técnicas propuestas y aprendiendo el cómo y porqué de muchas cosas vistas en clase de manera práctica.

Pasando ya a lo que respecta a los datos en sí, lo primero que observamos fue el alto nivel de desbalanceo en los mismos, hecho que tiene una serie de consecuencias importantes en el proyecto, no solo desde el punto de vista del modelado, sino también del análisis y tratamiento de los datos. Entre ellos podemos nombrar la necesidad de usar escalas logarítmicas a la hora de análisis gráficos, de estratificar si o si cuando separabamos los sets de datos, la aplicación de técnicas de oversampling y undersampling, etc. Todo esto para terminar desembocando en resultados que, según los objetivos y límites que determinara el área de negocio, pueden significar la necesidad de realizar numerosas investigaciones para detectar a los pocos fraudes existentes, algo en principio normal en proyectos de este estilo.

Más allá de este desbalanceo, los datos mostraban ciertas tendencias desde el análisis exploratorio, algunas más notorias que se terminaron cumpliendo, como lo eran la importancia de variables como Housing Status, Income o Credit Risk Score, mientras que otras sorprendieron para bien como para mal en importancia de variables. Como ejemplo de una de estas últimas, mediante la explicabilidad pudimos ver que el Payment Type no parecía ofrecer tanta información relevante para las decisiones del modelo como esperabamos, a pesar de haberle dado un tratamiento especial. Todo esto puede suceder y por eso no se trata de una ciencia totalmente exacta y cierta por el momento el hecho del entrenamiento de este tipo de modelos, debido a que hay mucho de prueba y error, además de que las relaciones son tan complejas que no es fácil predecirlas con el pensamiento humano simple, aunque el sentido común y el entendimiento de los datos ayude mucho a mejorarlos. Es por esto que la explicabilidad cada vez toma un valor mayor en el mundo del Data Science.

Una vez analizados los datos, el proceso de ingeniería y selección de variables también resultó indispensable, no solo porque sea la única manera de procesar los datos correctamente, sino porque el hacerlo de la manera adecuada ayuda a los modelos a entender mejor los inputs que recibe. Modelos que fueron comparados entre sí en base a las métricas más relevantes para los objetivos del modelo. En este caso se trataban de métricas relacionadas a la capacidad de captar verdaderos positivos, es decir, evaluamos la capacidad del modelo de otorgarle altas probabilidades de fraude a aplicaciones que realmente fueron fraudulentas. Por ello el F2 fue el score elegido para maximizar, ya que da mayor preponderancia a la detección de fraudes, pero sin dejar de lado que se necesita además contar con cierta precisión en el resultado final.

Una vez comparados los modelos, las elecciones podrían haber sido varias, pero en nuestro caso decidimos llevar adelante el modelo de boosting LightGBM por varias razones. Por un lado, fue el modelo de boosting que mejores métricas obtuvo sin hiperparametrizar, pero sobre todo debido a que obtiene resultados similares a los demás ensambladores en tiempos de ejecución y consumo de recursos bastante menor, lo cual lo hace una alternativa muy eficiente. Además, tiene una capacidad de mejora y complejización en base al hyperparameters tunning mayor a modelos que en su versión por default daban buenos resultados como el Gaussian Naive Bayes. Luego de buscar los mejores hiperparámetros, quedó resulto el problema de elección del mejor modelo y pasamos al principal problema que compete a este trabajo: La aplicación del modelo y medición final en nuestro set de test.

Esta aplicación del modelo sobre el set de test supuso el enfrentar al modelo entrenado a mediciones finales en datos "desconocidos". Los resultados no parecen ser los mejores a priori en niveles genéricos, pero sería muy erróneo de nuestra parte el medir algo sin saber todas las condiciones de uso en un negocio. Es por eso que se presentaron una serie de alternativas, las cuales dependen básicamente del umbral en el que decidimos etiquetar a un registro como fraude o no. Si ese umbral es más alto, encontraremos menos fraudes pero tampoco tendremos muchos usuarios a investigar debido a que fallaremos menos a la hora de decir que alguien puede ser un potencial defraudador. Lo mismo a la inversa, un umbral bajo permite detectar muchos fraudes pero genera mucho gasto en investigaciones de sujetos que probablemente no sean defraudadores.

En este sentido, por ejemplo una estrategia que busque maximizar nuestro score principal, el F2 (en 0.32), captaría un 47% de los fraudes, solo etiquetando mal a un 3% de los registros como fraudulentos. Por otro lado, una estratégia que requeriría mucho mayor presupuesto sería el buscar maximizar la cantidad de aciertos positivos, en donde se capturaría el 83% de los fraudes, pero casi el 20% de los usuarios totales deberían ser investigados. También existen puntos intermedios que podría determinar negocio, como límites a la cantidad de falsos fraudes que se está dispuesto a aceptar, limitando así el presupuesto en investigación con la mejor detección posible. Por ejemplo, si se decidiera que solo se aceptan un 5% de falsos positivos a investigar, se podría lograr una detección del fraude de aproximadamente el 55%.

Podríamos seguir con los ejemplos hasta el cansancio, pero la idea es mostrar que el modelo llega hasta ciertos límites a la hora de asignar probabilidades y que luego la aplicación y definición de los umbrales de tolerancia dependen de aspectos que no son del todo técnicos o matemáticos, sino también de las necesidades, objetivos y limitaciones del negocio en el que se esté aplicando, abriendo la posibilidad al uso de otras herramientas muy efectivas como lo son la matriz de costos por ejemplo.

Para cerrar con este apartado y el proyecto en sí, creo importante recalcar en primer lugar que realmente ha sido de gran utilidad para probar una gran cantidad de herramientas, entender muchos conceptos, aprender a codear de manera ordenada y encapsulada, también a gestionar un directorio de trabajo y su versionado, además de por supuesto comprender muchos detalles de todo el proceso de modelado. Aún así es importante recalcar que el hecho de elegir, tunear y aplicar un modelo no es todo el trabajo de un Data Scientist, sino que al contrario, el analizar y explotar el set de datos, todo lo que respecta a su preprocesamiento y por supuesto el posterior entendimiento y explicación de cómo funcionan los modelos también forman parte indispensable de un proceso muy amplio y extenso. Y que no termina siquiera aquí, sino que de él se desprende la última parte del proyecto, la cual espero continuar en una futura práctica, en donde la idea es culminar con la puesta en producción (al menos de manera simulada con Flask) de nuestro modelo aplicado.
